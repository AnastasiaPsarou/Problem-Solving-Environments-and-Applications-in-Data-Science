## <center> K-means <center> 

```python
#Importing required modules
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
 
pca = PCA(2)    
    
#Transform the data
df = pca.fit_transform(X)
 
df.shape
```

```python
#Import required module
from sklearn.cluster import KMeans
 
#Initialize the class object
kmeans = KMeans(n_clusters= 2)
 
#predict the labels of clusters.
label = kmeans.fit_predict(df)
 
print(label)
```

```python
from matplotlib import pyplot as plt

#Getting the Centroids
centroids = kmeans.cluster_centers_
u_labels = np.unique(label)
 
#plotting the results:
 
for i in u_labels:
    plt.scatter(df[label == i , 0] , df[label == i , 1] , label = i)
plt.legend()
plt.show()
```

```python
for i in u_labels:
    plt.scatter(df[label == i , 0] , df[label == i , 1] , label = i)
plt.scatter(centroids[:,0] , centroids[:,1] , s = 80, color = 'k')
plt.legend()
plt.show()
```

## <center> Elbow Method <center>

```python
kmeans5 = KMeans(n_clusters=5)
y_kmeans5 = kmeans5.fit_predict(X)
print(y_kmeans5)

kmeans5.cluster_centers_
```

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(X)
    kmeans.fit(X)
    Error.append(kmeans.inertia_)
    
    
import matplotlib.pyplot as plt
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

## <center> Hierarchical Clustering <center>

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.title("Hierarchical Clustering")

selected_data = X
clusters = shc.linkage(selected_data, 
            method='ward', 
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.show()
```

```python
u_labels = np.unique(clusters)
 
for i in u_labels:
    plt.scatter(df[label == i , 0] , df[label == i , 1])
plt.legend()
plt.show()
```

## <center> Conclusion <center>

The basic difference between the two methods is that in Kmeans the number of clusters must be predifined, while in hierarchical clustering the algorithm can stop at any number of clusters, one find appropriate by interpreting the dendrogram. Even the differences of the implemetations of the two algorithms we end up to similar results.
